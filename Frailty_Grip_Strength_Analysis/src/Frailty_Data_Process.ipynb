```python
import pandas as pd

# Load ingested data from Stage 1
df = pd.read_csv("/content/drive/MyDrive/Frailty_Grip_Strength_Analysis/Data_Clean/frailty_ingested.csv")

print(df)
```

```python
df = df.rename(columns={'Grip strength': 'Grip_kg'})
```

# **UNIT STANDARDIZATION**

```python
# PART a(i) - Convert Height from inches to metres

df['Height_m'] = df['Height'] * 0.0254
print(df[['Height', 'Height_m']])
```

```python
# PART a(ii) - Convert Weight from pounds to kilograms

df['Weight_kg'] = df['Weight'] * 0.45359237
print(df[['Weight', 'Weight_kg']])
```

# **FEATURE ENGINEERING**

```python
# PART b(i) - Calculate BMI using standardized units

df['BMI'] = (df['Weight_kg'] / (df['Height_m'] ** 2)).round(2)
print(df[['Height_m', 'Weight_kg', 'BMI']])
```

```python
# PART b(ii) - Assign AgeGroup category based on Age

age_bins   = [0, 29, 45, 60, 100]
age_labels = ["<30", "30–45", "46–60", ">60"]

df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
print(df[['Age', 'AgeGroup']])
```

# **Save Processed Data for Stage 3**

```python
df.to_csv("/content/drive/MyDrive/Frailty_Grip_Strength_Analysis/Data_Clean/frailty_processed.csv", index=False)
print("Stage 2 complete — data saved!")
```
