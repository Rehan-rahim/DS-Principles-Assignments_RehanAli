```python
import pandas as pd

# Load ingested data from Stage 1
df = pd.read_csv("/content/drive/MyDrive/Student_Performance_Analysis/Data_Clean/std_Performance_ingested.csv")
```

```python
# Check for missing values in each column
print(df.isnull().sum())
```

```python
# Check data types of each column
print(df.dtypes)
```

```python
# Check for duplicate rows
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Check basic statistics for score columns
print(df.describe())
```

```python
# Add overall average score column

df['overall_avg'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
df['overall_avg'] = df['overall_avg'].round(2)

print(df[['math score', 'reading score', 'writing score', 'overall_avg']].head())
```

```python
# Save processed data for Stage 3

df.to_csv("/content/drive/MyDrive/Student_Performance_Analysis/Data_Clean/std_Performance_processed.csv", index=False)
print("Stage 2 complete — data saved!")
```
